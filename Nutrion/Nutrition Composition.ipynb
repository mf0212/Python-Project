## Project: Nutrition Data Visualization on 80 cereal products

Link dataset : https://www.kaggle.com/crawford/80-cereals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset
cereal_df = pd.read_csv("data/cereal.csv")
```

```python
cereal_df.head()
```

```python
#Irrelevant fields
fields = ['shelf', 'weight', 'cups', 'rating']

cereal_df_new = cereal_df.drop(columns = fields)
cereal_df_new.head()
```

```python
#corr()

cereal_corr = cereal_df_new.corr() # Get correlation data
cereal_corr
```

```python
#ones_like can build a matrix of booleans(True, False) with the same shape as our data

ones_corr = np.ones_like(cereal_corr, dtype = bool)
ones_corr
```

```python
ones_corr.shape,cereal_corr.shape
```

```python
#np's triu : return only upper triangle matrix

mask = np.triu(ones_corr)
mask
```

```python
adjusted_mask = mask[1:,:-1]
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:,:-1]
adjusted_cereal_corr
```

```python
fig , ax = plt.subplots(figsize = (10,8))

cmap = sns.diverging_palette(0,230,90,60, as_cmap = True)
sns.heatmap(data = adjusted_cereal_corr, mask = adjusted_mask
            , annot = True, annot_kws = {'fontsize':10},fmt = '.2f',cmap="Blues"
           , vmin = -1, vmax = 1, linecolor = "white", linewidths= 1)

yticks = [i.capitalize() for i in adjusted_cereal_corr.index]
xticks = [i.capitalize() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks,rotation = 0);
ax.set_xticklabels(xticks,rotation = 0);
ax.set_title("Correlation matrix\nSampled cereals compostion\n", loc = 'left', fontsize = 10);
```
