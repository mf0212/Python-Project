```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.formula1.com/en/results.html/2023/drivers.html"
respone = requests.get(url)
```

```python
if respone.status_code == 200 : # Doc thanh cong
    soup = BeautifulSoup(respone.content,'html.parser')
    table = soup.find("table", class_ = "resultsarchive-table")
    # Extract the row of the table
    rows = table.find_all('tr')
    for row_id,row in enumerate(rows):
        columns = row.find_all("td")
        # print(f"row = {row_id} : {columns}")
        if columns:
            positions = columns[1].text.strip()
            drive_name = columns[2].text.strip().split("\n")
            drive_name = " ".join(drive_name[:-1])
            nationality = columns[3].text.strip()
            car = columns[4].text.strip()
            pts = columns[5].text.strip()
            print(f"Position: {positions}\
                  \nDrive Name: {drive_name}\
                  \nNationality: {nationality}\
                  \nCar: {car}\
                  \nPoint: {pts}\n")
```
